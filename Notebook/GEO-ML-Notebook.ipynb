### GEO-MACHINE LEARNING

```python
import geopandas as gpd
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# 1. Load data
gdf = df

# 2. Clean target
gdf = gdf[gdf["fclass"].notna()].copy()

# 3. Basic cleaning
for col in ["name", "ref"]:
    if col in gdf.columns:
        gdf[col] = gdf[col].fillna("unknown")

def to_binary(series):
    s = series.astype(str).str.lower().str.strip()
    return s.map({
        "yes": 1, "y": 1, "true": 1, "1": 1, "t": 1,
        "no": 0, "n": 0, "false": 0, "0": 0, "f": 0
    }).fillna(0).astype(int)

for col in ["oneway", "bridge", "tunnel"]:
    if col in gdf.columns:
        gdf[col] = to_binary(gdf[col])


if "maxspeed" in gdf.columns:
    gdf["maxspeed"] = (
        gdf["maxspeed"]
        .astype(str)
        .str.extract(r"(\d+)", expand=False)
        .astype(float)
    )

if "layer" in gdf.columns:
    gdf["layer"] = pd.to_numeric(gdf["layer"], errors="coerce").fillna(0)

# 4. Geometry features
gdf["length_m"] = gdf.geometry.length
gdf["num_vertices"] = gdf.geometry.apply(lambda x: len(x.coords) if x else 0)
gdf["start_x"] = gdf.geometry.apply(lambda x: x.coords[0][0] if x else np.nan)
gdf["start_y"] = gdf.geometry.apply(lambda x: x.coords[0][1] if x else np.nan)
gdf["end_x"] = gdf.geometry.apply(lambda x: x.coords[-1][0] if x else np.nan)
gdf["end_y"] = gdf.geometry.apply(lambda x: x.coords[-1][1] if x else np.nan)

# 5. Target
y = gdf["fclass"].astype(str)

# 6. Feature columns
feature_cols = [
    "code", "name", "ref", "oneway", "maxspeed", "layer",
    "bridge", "tunnel", "length_m", "num_vertices",
    "start_x", "start_y", "end_x", "end_y"
]

feature_cols = [c for c in feature_cols if c in gdf.columns]
X = gdf[feature_cols].copy()

# 7. One-hot encode categoricals
categorical_cols = [c for c in ["code", "name", "ref"] if c in X.columns]
numeric_cols = [c for c in X.columns if c not in categorical_cols]

X = pd.get_dummies(X, columns=categorical_cols, dummy_na=False)

# 8. Final cleanup
X = X.replace([np.inf, -np.inf], np.nan).fillna(0)

print(X.shape, y.shape)
```
